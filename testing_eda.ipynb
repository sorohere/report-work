```python
# Import necessary libraries  
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns  
```

```python
# Set up global plot styling  
plt.rcParams['font.family'] = 'sans-serif'  
plt.rcParams['font.sans-serif'] = ['Lato', 'IBM Plex Sans', 'Arial']  
  
# Read the data from paste1.txt (assuming it has comma-separated values)  
df = pd.read_csv('data.csv')  
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Filter out rows with missing 'tenure' and 'churn'
scatter_df = df.dropna(subset=['tenure', 'churn'])

# Create jitter for the 'churn' column to avoid overlap in plotting
jitter_scatter = np.random.normal(0, 0.04, size=len(scatter_df))  # slightly reduced jitter
churn_jittered = scatter_df['churn'] + jitter_scatter

# Create the figure and axis for scatter plot
fig, ax1 = plt.subplots(figsize=(10, 6))  # wider figure
plt.subplots_adjust(left=0.25, right=0.9, top=0.88, bottom=0.15)  # more space on the left

# Plot the scatter plot with jitter (using a color map)
scatter = ax1.scatter(scatter_df['tenure'], churn_jittered,
                      alpha=0.6, c=scatter_df['churn'], cmap='coolwarm',
                      s=50, edgecolor='#333333', linewidth=0.5)

# Set title and labels
ax1.set_title('Relationship Between Tenure and Churn', fontsize=20, fontweight='semibold', color='#222222', pad=15)
ax1.set_xlabel('Tenure (months)', fontsize=16, fontweight='medium', color='#333333', labelpad=10)
ax1.set_ylabel('Churn Status (with jitter)', fontsize=16, fontweight='medium', color='#333333', labelpad=12)

# Configure ticks and grid lines
ax1.tick_params(axis='both', labelsize=14, colors='#555555')
ax1.grid(True, linestyle='--', alpha=0.7, color='#E0E0E0')

# Set y-axis limits and ticks to clearly label churn statuses
ax1.set_ylim(-0.2, 1.2)
ax1.set_yticks([0, 1])
ax1.set_yticklabels(['Not Churned', 'Churned'], fontsize=14)

# Add horizontal reference lines
ax1.hlines(y=[0, 1], xmin=scatter_df['tenure'].min(), xmax=scatter_df['tenure'].max(),
           colors='gray', linestyles='dotted', alpha=0.4)

# Style the spines
for spine in ax1.spines.values():
    spine.set_color('#333333')
    spine.set_linewidth(0.8)

# Show the scatter plot
plt.tight_layout()
plt.show()
```

```python
from matplotlib.ticker import MaxNLocator
import seaborn as sns
import matplotlib.pyplot as plt

boxplot_df = df.dropna(subset=['rev_per_month', 'account_segment'])

fig, ax2 = plt.subplots(figsize=(9, 6))
plt.subplots_adjust(left=0.25, right=0.85, top=0.85, bottom=0.15)
ax2.set_axisbelow(True)

sns.boxplot(x='account_segment', y='rev_per_month', data=boxplot_df, ax=ax2,
            palette=['#766CDB', '#DA847C', '#D9CC8B', '#7CD9A5', '#877877', '#52515E'])

ax2.set_title('Revenue per Month by Account Segment', fontsize=20, fontweight='semibold', color='#222222', pad=15)
ax2.set_xlabel('Account Segment', fontsize=16, fontweight='medium', color='#333333', labelpad=10)
ax2.set_ylabel('Revenue per Month', fontsize=16, fontweight='medium', color='#333333', labelpad=12)

# Apply better formatting to avoid overlap
ax2.yaxis.set_major_locator(MaxNLocator(nbins=10))
plt.setp(ax2.get_xticklabels(), rotation=15, ha='right')  # Optional: for x-axis clarity
ax2.tick_params(axis='y', labelsize=10)

ax2.grid(True, linestyle='--', alpha=0.7, color='#E0E0E0')
for spine in ax2.spines.values():
    spine.set_color('#333333')
    spine.set_linewidth(0.8)

plt.tight_layout()
plt.show()
```
